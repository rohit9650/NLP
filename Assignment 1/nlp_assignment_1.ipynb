```
'''
Utilities Functions
'''

def substiute(term_list, substitutions):
  output_length = ""
  for word in term_list:
      if word.lower() in substitutions.keys():
        output_length = output_length + " " + substitutions[word.lower()]
      else:
        output_length = output_length + " " + word
  return output_length

def standardizeSpeed(text):
    
    substitutions={"metres per second": "l/t",
                   "metres per hour": "l/t",
                   "metres per minute": "l/t",
                   "kilometres per second": "l/t",
                   "kilometres per hour": "l/t",
                   "miles per second": "l/t",
                   "miles per hour": "l/t",
                   "miles per minute": "l/t",
                   "konts": "l/t",
                   "feet per second": "l/t",
                   "km/hr": "l/t",
                   "km/h": "l/t",
                   "m/s": "l/t",
                   "mi/h": "l/t",
                   "kmh": "l/t",
                   "metres per second.": "l/t",
                   "metres per hour.": "l/t",
                   "metres per minute.": "l/t",
                   "kilometres per second.": "l/t",
                   "kilometres per hour.": "l/t",
                   "miles per second.": "l/t",
                   "miles per hour.": "l/t",
                   "miles per minute.": "l/t",
                   "konts.": "l/t",
                   "feet per second.": "l/t",
                   "km/hr.": "l/t",
                   "km/h.": "l/t",
                   "m/s.": "l/t",
                   "mi/h.": "l/t",
                   "kmh.": "l/t"
                   }
    
    term_list = word_tokenize(text)
    output_length = substiute(term_list, substitutions)
    
    return output_length

def standardizeTime(text):
    
    substitutions={"seconds": "t",
                   "minutes": "t",
                   "hours": "t",
                   "second": "t",
                   "minute": "t",
                   "min": "t",
                   "hour": "t",
                   "s": "t",
                   "seconds.": "t",
                   "minutes.": "t",
                   "hours.": "t",
                   "second.": "t",
                   "minute.": "t",
                   "min.": "t",
                   "hour.": "t",
                   "s.": "t"
                   }
    term_list = word_tokenize(text)
    output_length = substiute(term_list, substitutions)
    
    return output_length

def standardizeLength(text):
    substitutions={"metres": "l",
                   "miles": "l",
                   "degree": "l",
                   "meters": "l",
                   "miles": "l",
                   "kilometres": "l",
                   "centimetres": "l",
                   "ms": "l",
                   "mi": "l",
                   "m": "l",
                   "cms" :"l",
                   "kms" :"l",
                   "m": "l",
                   "cm": "l",
                   "km": "l",
                   "metres.": "l",
                   "miles.": "l",
                   "degree.": "l",
                   "meters.": "l",
                   "miles.": "l",
                   "kilometres.": "l",
                   "centimetres.": "l",
                   "ms.": "l",
                   "mi.": "l",
                   "m.": "l",
                   "cms." :"l",
                   "kms." :"l",
                   "m.": "l",
                   "cm.": "l",
                   "km.": "l"
                   }
    term_list = word_tokenize(text)
    output_length = substiute(term_list, substitutions)
    
    return output_length

def standardizeAcc(text):
    substitutions={"m/s2": "l/tsq",
                   "m/s^2": "l/tsq",
                   "metres per second square": "l/tsq",
                   "metres per sqaure second": "l/tsq",
                   "m/s2.": "l/tsq",
                   "m/s^2.": "l/tsq",
                   "metres per second square.": "l/tsq",
                   "metres per sqaure second.": "l/tsq"
                   }
    
    term_list = word_tokenize(text)
    output_length = substiute(term_list, substitutions)
    
    return output_length
 
```

```
def standardizeForm(text):
    
  accleration = standardizeAcc(text)
  speed = standardizeSpeed(accleration)
  time_speed = standardizeTime(speed)
  length_time_speed =  standardizeLength(time_speed)
  
  return length_time_speed
```

```
distance = {'distance':"distance",
            'displacement':"displacement",
            'area':"area",
            'height':"height",
            'far':"far",
            'tall':"tall",
            'length':"length",
            'orbit':"orbit",
            'high':"high",
            'angle':"angle",
            'altitude':"altitude",
            'radius':"radius",
            'scope':"scope",
            'seperation':"seperation",
            'size':"size",
            'stretch':"stretch",
            'width':"width",
            'degreee':"degree",
            'position':"position"
        }
speed = {'velocity':"velocity",
         'agility':"agility",
         'pace':"pace",
         'quickness':"quickness",
         'swiftness':"swiftnes",
         'speed':"speed",
         'fast':"fast"
         }
time = {'time':"time",
        'long':"long",
        'seconds':"seconds",
        'hang time':"hang time",
        'hangtime':"hangtime",
        'hang-time':"hang-time",
        'in the air':"in the air"
        }

accleration = {'acceleration':"acceleration",
               'deceleration':"deceleration",
               'upward':"acceleration",
               'downward':"acceleration",
               'retardation':"acceleration",
               'vertically':"acceleration",
               'upwards':"acceleration",
               'downwards':"acceleration",
               'accelerated':"acceleration",
               'accelerating':"acceleration"
               }

def extractInformation(sentence):
    information = {'Speed_as_value':0,
                   'Distance_as_value':0,
                   'Acceleration_as_value':0,
                   'Time_as_value':0,
                   'Speed_as_ques':0,
                   'Distance_as_ques':0,
                   'Acceleration_as_ques':0,
                   'Time_as_ques':0
                   }
    term_list = word_tokenize(sentence)
    post_process = term_list
    
    for word in term_list:
        if word == "l/t":
            information['Speed_as_value'] = 1
            i = 0
            while i < len(post_process):
              w = post_process[i]
              if w.lower() in speed.values():
                del post_process[i]
                break
              else:
                i = i + 1 
        if word == "l":
            information['Distance_as_value'] = 1
            i = 0
            while i < len(post_process):
              w = post_process[i]
              if w.lower() in distance.values():
                del post_process[i]
                break
              else:
                i = i + 1
        if word == "t":
            information['Time_as_value'] = 1
            i = 0
            while i < len(post_process):
              w = post_process[i]
              if w.lower() in time.values():
                del post_process[i]
                break
              else:
                i = i + 1
        if word == "l/tsq":
            information['Acceleration_as_value'] = 1
            i = 0
            while i < len(post_process):
              w = post_process[i]
              if w.lower() in accleration.values():
                del post_process[i]
                break
              else:
                i = i + 1
   
    for i in range(len(post_process)):
        word = post_process[i]
        if word.lower() in speed.values():
            information['Speed_as_ques'] = 1
        if word.lower() in distance.values():
            information['Distance_as_ques'] = 1
        if word.lower() in time.values():
            information['Time_as_ques'] = 1
        if word.lower() in accleration.values():
            information['Acceleration_as_ques'] = 1
    return information
```

```
def splitNumbersAndUnits(text):
  term_list = word_tokenize(text)
  new_term = ""
  change_terms = {}
  problem = []
  pos = 0
  for term in term_list:
    if term[0].isdigit():
      problem.append(term)
  for word in problem:
    for i in range(0,len(word)):
      if not word[i].isdigit():
        pos = i
        break
    new_term = word[:pos] + " " + word[pos:]
    change_terms[word] = new_term
  new_text = ""
  for term in term_list:
      if term in change_terms.keys():
          new_text = new_text + " " + change_terms[term]
      else:
          new_text = new_text + " " + term
  return new_text
```

```
def get_files(dir_path):
  """
  return the list of all files present in dir_path
  """
  files = []
  for file in sorted(os.listdir(dir_path)):
      files.append(file)
  return files
```

```
def process_corpus(dir_path):
  '''
  process corpus file by file to form a dataframe (Binary Incidence Matrix)
  input : directory path of corpus
  '''
  files = get_files(dir_path)
  index = 0
  for file in files:
      file_path = dir_path + "/" + file
      fp = open(file_path,"r")
      file_content = fp.read()
      
      pre_processed = splitNumbersAndUnits(file_content)
      std_text = standardizeForm(pre_processed)
      print(std_text)
      quantities = extractInformation(std_text)
      print(quantities)
      column_entries = list(quantities.values())
      
      column_name = "D" + str(index)
      df_[column_name] = column_entries
      index = index + 1
      print("\n\n\n")
process_corpus('/content/corpus')   
```
